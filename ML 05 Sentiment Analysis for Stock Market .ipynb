## train test split 

```python
x_scores = sentiment_data.iloc[:,0:7]
y_scores = sentiment_data.iloc[:,7] 

x_train, x_test, y_train, y_test = train_test_split(x_scores, 
                                                    y_scores, 
                                                    test_size=0.30,
                                                    random_state=1910)

```

```python
class fun_t(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.times = []

    def on_epoch_begin(self, batch, logs={}):
        self.epoch_time_start = time.time()

    def on_epoch_end(self, batch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)  
```

```python
def dataset(xtra,x,y,time_steps):
  import numpy as np
  xs, ys = [], []
  xtra=xtra
  xnum=xtra.to_numpy()
  for i in range(len(x) - time_steps):
    v = x.iloc[i : (i+time_steps)].to_numpy()
    comb=np.concatenate((v, xnum[i+time_steps]), axis=None)
    xs.append(comb)
    ys.append(y.iloc[i+time_steps]) #So that next y is considered
  return np.array(xs), np.array(ys)  
```

## create time lags for time series prediction 


```python
TIME_STEPS = 1 
```

```python
x_train,y_train = dataset(x_train,y_train,y_train,TIME_STEPS) 
```

```python
x_test,y_test = dataset(x_test,y_test,y_test,TIME_STEPS) 
```

```python
x_train=x_train.reshape((x_train.shape[0],1,x_train.shape[1]))
x_test=x_test.reshape((x_test.shape[0],1,x_test.shape[1])) 
```

## LSTM model 

```python
np.random.seed = 1910
tf.random.set_seed = 1910

time3=fun_t()
history=History()
lstm_model = Sequential()
lstm_model.add(keras.layers.LSTM(units = 1, input_shape = (x_train.shape[1],x_train.shape[2])))
lstm_model.add(keras.layers.Dense(units = 1))
lstm_model.compile(loss='mean_squared_error', optimizer='adam') 
history = lstm_model.fit(x_train,y_train, epochs= 200, shuffle = False, batch_size = 10, callbacks=[history,time3],validation_split = 0.3,verbose = 0)
```

```python
time_lstm = list(time3.times)
print('Long Short Term Memory Model Time Taken: ',sum(time_lstm))
train_lstm = history.history.get('loss')
val_lstm = history.history.get('val_loss')

base_mse=val_lstm[-1]
print('Base MSE for the above model:',base_mse)
print('Base Accuracy for the above model:',1-base_mse) 
```
